```python
#find total number of duplicates to remove 
sum(dataframe.duplicated())
```

```python
#check if any values are null
dataframe.isnull()
```

```python
#remove the rows where all the column values are null
d1 = dataframe.dropna(how="all")
d1.info()
```

```python
#drop rows based on a threshold
d2=dataframe.dropna(thresh=0.8)
d2.info()
```

```python
dataframe.head(20)
```

```python
# assigning mean value to missing values in trip_miles column
dataframe['trip_miles'].fillna(dataframe['trip_miles'].mean(),inplace=True)
trip_miles = pd.DataFrame({'trip_miles': dataframe['trip_miles']})
trip_miles.head()
```

```python
# assigning mean value to missing values in the trip_total column

dataframe['trip_total'].fillna(dataframe['trip_total'].mean(),inplace=True)
trip_total = pd.DataFrame({'trip_total': dataframe['trip_total']})
trip_total.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#feature variable
X = trip_miles

#target variable
y = trip_total

#splitting data into training and testing set
X_train, X_test , y_train , y_test= train_test_split(X,y,test_size=0.2,randon_state=23)

#training linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
#prediction
y_predict = model.predict(X_train)
y_predict
```

```python
import matplotlib.pyplot as pyt

pyt.scatter(X_train,y_train)
pyt.plot(X_train,y_predict, color= 'yellow')
pyt.xlabel('trip miles')
pyt.ylabel('trip total')
```

```python
y_predict_test = model.predict(X_test)

pyt.scatter(X_test,y_test)
pyt.plot(X_test, y_predict_test, color = 'red')
pyt.xlabel('trip miles')
pyt.ylabel('trip total')
```
